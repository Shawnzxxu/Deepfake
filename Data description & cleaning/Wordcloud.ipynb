```python
import json
import stylecloud

file_path = '/content/drive/MyDrive/deepfake/Processed_deepfake.json'
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

body_texts = [item['body'] for item in data]
all_text = ' '.join(body_texts)

stopwords = {'people', 'like', 'would', 'one', 'make', 'get', 'think', 'could', 'it', 'going', 'really'}

stylecloud.gen_stylecloud(
    text=all_text,
    icon_name='fab fa-reddit-square',
    palette='colorbrewer.qualitative.Dark2_8',
    background_color='white',
    size=1024,
    output_name='stylecloud_output.png',
    custom_stopwords=stopwords,
    max_font_size=200,
    random_state=333
)

from IPython.display import Image
Image(filename='stylecloud_output.png')
```
